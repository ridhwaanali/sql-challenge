```python
#Employee Data
employee_path= Path("employees.csv")
employee_data= pd.read_csv(employee_path)
employee= pd.DataFrame(employee_data)
employee.head()
```

```python
#Department Data
department_path= Path("departments.csv")
department_data= pd.read_csv(department_path)
department= pd.DataFrame(department_data)
department.head()
```

```python
#Title Data
title_path= Path("titles.csv")
title_data= pd.read_csv(title_path)
title= pd.DataFrame(title_data)
title.head()
```

```python
#Department Employee Data
dept_emp_path= Path("dept_emp.csv")
dept_emp_data= pd.read_csv(dept_emp_path)
dept_emp= pd.DataFrame(dept_emp_data)
dept_emp.head()
```

```python
#Salary Data
salary_path= Path("salaries.csv")
salary_data= pd.read_csv(salary_path)
salary= pd.DataFrame(salary_data)
salary.head()
```

```python
#Department Management Data
dept_manager_path= Path("dept_manager.csv")
dept_manager_data= pd.read_csv(dept_manager_path)
dept_manager= pd.DataFrame(dept_manager_data)
dept_manager.head()
```

```python
#Merging Employee and Salary to get employee number, first and last name, sex, and salary
employee_salary_merge= pd.merge(employee, salary, on= 'emp_no')
employee_salary_merge.head()
```

```python
#Dropping the columns that are needed
employee_salary= employee_salary_merge.drop(columns=['emp_title_id', 'birth_date', 'hire_date'])
employee_salary.head()
```

```python
#Modifying employee data to find first name, last name, hire dates
employee_hired= employee.drop(columns=['emp_no', 'emp_title_id', 'birth_date', 'sex'])
employee_hired.head()
```

```python
# employees hired in 1986
employee_1986= employee_hired.loc(employee_hired['hire_date'] >= '01/01/1986') & (employee_hired['hire_date'] < '01/01/1987')
employee_1986
```

```python
#Listing manager of each department by merging department, dept_emp, employee
department_manager_merge= pd.merge(department,dept_emp)
department_manager= pd.merge(department_manager_merge, employee)
department_manager.head()
```

```python
#cleaning data to focus on mainly dept_no, dept_name, emp_no, first_name, last_name
department_manager_cleaned= department_manager.drop(columns=['emp_title_id', 'birth_date', 'sex', 'hire_date'])
department_manager_cleaned.head()
```
